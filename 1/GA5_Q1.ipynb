```python
# K-Means Clustering on Product Embeddings - FREE VERSION (No API Key)
# This uses a FREE local embedding model instead of OpenAI
# ⚠️ WARNING: This will give DIFFERENT results than the official OpenAI version
# Use this to learn the process, but you'll need OpenAI API for the real submission

# Install required packages (free!)
!pip install sentence-transformers scikit-learn numpy

from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans
import numpy as np

# Product descriptions (all 50)
descriptions = """Digital kitchen scale with tare function and precise 1-gram increments for baking recipes
Resistance loop mini bands for hip abductor activation and warm-up mobility routines
Bamboo plant stakes with soft tie tape for supporting tall tomato and pepper plants
Compression knee sleeve with patella support for running, cycling, and weightlifting safety
Acrylic document tray organiser with three tiered slots for desktop paper management
Wireless noise-cancelling headphones with 30-hour battery life and multipoint Bluetooth pairing
Wooden cutting board with juice groove and rubber feet for slicing bread and vegetables
Agility ladder with adjustable rungs for speed, coordination, and footwork training drills
Portable SSD with 2TB storage and USB 3.2 Gen 2 transfer speeds for content creators
Stainless-steel spice rack with labelled magnetic tins for organised pantry storage
Mortar and pestle set in granite for grinding fresh spices and making pesto from scratch
Stand mixer with dough hook, whisk, and paddle attachments for bread and cake making
Earthworm castings premium compost tea for boosting microbial activity in grow bags
Stainless-steel pruning shears with ergonomic grip and sap groove for clean stem cuts
Ultra-thin 4K OLED laptop with AI-powered processor and backlit mechanical keyboard
Compost aerator tool with stainless-steel shaft for turning and oxygenating garden compost
Biodegradable seed trays with moisture-retaining peat plugs for easy seedling transplanting
Silicone baking mat set with measurement guides for macarons and pastry sheet rolling
Automatic greenhouse vent opener with heat-sensitive actuator for temperature regulation
Foldable wireless charging pad compatible with all Qi-enabled smartphones and earbuds
Expandable garden hose with brass fittings and eight-pattern spray nozzle for watering
Handheld milk frother with three speed settings for lattes, cappuccinos, and matcha drinks
High-refresh-rate gaming monitor with adaptive sync and ultra-wide curved display panel
Oven-safe ramekins set for individual souffles, creme brulee, and egg en cocotte dishes
Organic slow-release fertiliser granules enriched with nitrogen for lush vegetable patches
Smart home security camera with night vision, motion detection, and two-way audio
Calligraphy practice set with guide sheets, oblique pen holder, and four nib sizes
Hardcover recipe journal with tabbed sections for organising family cookbook recipes
Voice-assistant-enabled Bluetooth speaker with 360-degree surround sound and LED display
Foam roller with deep-tissue ridges for post-workout muscle recovery and fascial release
Stainless-steel pressure cooker with multiple safety valves for fast bean and lentil cooking
Copper-coated slug deterrent tape to protect seedlings in raised beds and containers
Professional-grade USB microphone with cardioid pattern for podcasting and live streaming
Wildflower seed mix blend for creating pollinator-friendly meadows in open garden borders
Compact e-reader with glare-free display and weeks-long battery life for avid readers
360-degree action camera with image stabilisation and 8K video recording capability
Extra-thick non-slip yoga mat with alignment guides for hot yoga and pilates practice
Adjustable hex dumbbell set with quick-lock collar for compact home strength training
High-carbon stainless-steel chef's knife with full tang and ergonomic pakkawood handle
Balance board trainer with anti-slip surface for core stability and rehabilitation exercises
Ergonomic wireless mouse with adjustable DPI and silent click switches for remote work
Sustainable mulch mat rolls to suppress weeds and retain soil moisture around shrubs
Heart rate monitor chest strap with Bluetooth for accurate calorie tracking during workouts
Professional stapler with flat-clinch mechanism and two-hundred sheet capacity for offices
Wi-Fi 6E mesh router system providing whole-home gigabit coverage with low latency
Smartwatch with ECG sensor, GPS tracking, and sleep analysis for health monitoring
Set of pastel highlighters in six shades for colour-coding textbook notes and revision
Mechanical keyboard with hot-swappable switches and per-key RGB lighting for gaming
Frost-protection fleece wrap for overwintering delicate perennials and citrus pot plants
Silicone spatula set heat-resistant to 600 degrees Fahrenheit for stirring and folding""".strip().split('\n')

print(f"✅ Loaded {len(descriptions)} product descriptions")

# Load FREE local embedding model (no API needed!)
print("\n📥 Loading embedding model (this may take a minute first time)...")
model = SentenceTransformer('all-MiniLM-L6-v2')  # Small, fast, free model
print("✅ Model loaded!")

# Generate embeddings (runs locally, completely free!)
print("\n🔄 Generating embeddings locally...")
embeddings = model.encode(descriptions, show_progress_bar=True)
embedding_matrix = np.array(embeddings)

print(f"✅ Embedding matrix shape: {embedding_matrix.shape}")

# Run K-Means clustering with SAME parameters as assignment
print("\n🎯 Running K-Means clustering...")
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(embedding_matrix)

# Count items in each cluster
unique_labels, counts = np.unique(cluster_labels, return_counts=True)

print("\n" + "="*60)
print("📊 CLUSTER DISTRIBUTION:")
print("="*60)
for label, count in zip(unique_labels, counts):
    print(f"Cluster {label}: {count:2d} items")

# Find cluster with most items
max_cluster_idx = np.argmax(counts)
max_cluster_label = unique_labels[max_cluster_idx]
max_cluster_count = counts[max_cluster_idx]

# Output result
result = f"{max_cluster_label}, {max_cluster_count}"
print("\n" + "="*60)
print(f"🎯 ANSWER (FREE MODEL): {result}")
print("="*60)

print("\n⚠️  NOTE: This answer is from a FREE local model.")
print("    Your assignment requires OpenAI's text-embedding-3-small,")
print("    which will give a DIFFERENT answer.")
print("    This is for learning only - you still need OpenAI API for submission!")

# Optional: Show which products are in each cluster
print("\n" + "="*60)
print("📋 CLUSTER BREAKDOWN (what's in each group):")
print("="*60)
for cluster_id in unique_labels:
    indices = np.where(cluster_labels == cluster_id)[0]
    print(f"\n🔹 Cluster {cluster_id} ({len(indices)} items):")
    for idx in indices[:3]:  # Show first 3 items
        print(f"   • {descriptions[idx][:70]}...")
    if len(indices) > 3:
        print(f"   ... and {len(indices)-3} more")
```
